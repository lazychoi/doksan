```python
import time
import requests
from bs4 import BeautifulSoup
import codecs
```

```python
res = requests.get('https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&start=we')
res
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
title_html = soup.select('a.bo3')
titles = []
for item in title_html:
    titles.append(item.text)
titles, len(titles)
```

```python
author_html = soup.select('div.ss_book_list > ul > li:nth-child(3)')

authors = []
for item in author_html:
    line = item.text
    if '지은이' in line or '엮은이' in line:
        authors.append( line.split(' | ')[0] )
    
authors, len(authors)
```

```python
price_html = soup.select('.ss_p2 span')

prices = []
for item in price_html:
    prices.append(item.text)
prices, len(prices)
```

```python
import os
path = os.getcwd()  # 현재 작업 디렉토리 가져오기

bestseller_50 = zip(titles, authors, prices)

with codecs.open(f"{path}/aladin.txt", mode='w') as f:

    for item in bestseller_50:
        f.write(str(item)[1:-1])
        f.write('\n')

with codecs.open(f"{path}/aladin.txt", mode='r') as f:
    print( f.readlines() )
```
